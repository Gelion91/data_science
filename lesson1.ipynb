```python
import csv
```

```python
streets = {}
```

```python
%time
with open('data-398-2019-12-31.csv', 'r', encoding='cp1251') as file:
    #fields = ['ID', 'Name', 'Longitude_WGS84', 'Latitude_WGS84', 'Street', 'AdmArea', 'District', 'RouteNumbers',
    #         'StationName', 'Direction', 'Pavilion', 'OperatingOrgName', 'EntryState', 'global_id', 'geoData']
    reader = csv.DictReader(file, delimiter=';')
    for row in reader:
        if row['Street'] in streets:
            streets[row['Street']] += 1
        else:
            streets[row['Street']] = 1
```
